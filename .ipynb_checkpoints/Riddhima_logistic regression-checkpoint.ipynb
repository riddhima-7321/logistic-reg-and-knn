```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df.loc[ :,'gender'] = label_encoder.fit_transform(df['gender'])
df.loc[ :,'head_gender'] = label_encoder.fit_transform(df['head_gender'])
# df.head()
```

```python
df['industry']=label_encoder.fit_transform(df['industry'])
# df[30:]
```

```python
df['profession']=label_encoder.fit_transform(df['profession'])
# df[30:]
```

```python
df['greywage'] = label_encoder.fit_transform(df['greywage'])
df['coach'] = label_encoder.fit_transform(df['coach'])
df['way'] = label_encoder.fit_transform(df['way'])
df['traffic'] = label_encoder.fit_transform(df['traffic'])
```

```python
col_drop=['event']
X = df.drop(columns=col_drop, axis=1)
y=df['event']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.30,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
classify=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1', 'l2', 'elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classsify_regress=GridSearchCV(classify,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classsify_regress.fit(X_train,y_train)
```

```python
print(classsify_regress.best_params_)
```

```python
print(classsify_regress.best_score_)
```

```python
y_pred=classsify_regress.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
score
```
