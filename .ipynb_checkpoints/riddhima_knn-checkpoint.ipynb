```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('Type',axis=1))
```

```python
scaled_features=scaler.transform(df.drop('Type',axis=1))
```

```python
df_feat=pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,y_train,y_test=train_test_split(scaled_features,df['Type'],test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
accuracy_rate=[]
for i in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,df_feat,df['Type'],cv=9)
    accuracy_rate.append(score.mean())
```

```python
plt.plot(range(1,50),accuracy_rate,linestyle='dashed',marker='o',markerfacecolor='red')
```

```python
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(pred,y_test)
score
```
